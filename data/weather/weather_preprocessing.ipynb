```python
import pandas as pd

#load the weather dataset and the labels
weather_data = pd.read_csv('weather_data.csv')
weather_labels = pd.read_csv('weather_data_labels.csv')

# Display basic information about the dataset
print(weather_data.info())
print(weather_labels.head())
print(weather_data.head())
```

```python
print(weather_data.isnull().sum())
```

```python
merged_data = pd.merge(weather_data, weather_labels, on='DATE')
merged_data.head()
```

```python
merged_data
```

```python
picnic_weather_columns = [col for col in merged_data.columns if "_picnic_weather" in col]
picnic_weather_columns
```

```python
import os

# Create a directory to save the CSV files
os.makedirs('city_data', exist_ok=True)

# Filter and save each city's data into separate CSV files
for col in picnic_weather_columns:
    city_name = col.replace('_picnic_weather', '')
    city_cols = merged_data.filter(like=city_name).columns
    city_data = merged_data[['DATE'] + city_cols.tolist()]
    file_name = f'city_data/{city_name}.csv'
    city_data.to_csv(file_name, index=False)
    print(f"Saved data for {city_name} to {file_name}")
```

```python
# correlation matrix for one of the cities
import seaborn as sns
import matplotlib.pyplot as plt

city_name = 'STOCKHOLM'
city_cols = merged_data.filter(like=city_name).columns
city_data = merged_data[['DATE'] + city_cols.tolist()]

corr = city_data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'Correlation Matrix for {city_name}')
plt.show()
```
